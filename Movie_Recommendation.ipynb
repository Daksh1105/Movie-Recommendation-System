Importing libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Preprocessing

```python
movies_data = pd.read_csv('/content/movies.csv')
movies_data.head()
```

```python
movies_data.shape
```

```python
#feature extraction
selected_features = ['genres','keywords','tagline', 'cast','director',]
print(selected_features)
```

```python
#replacing null values with null strings
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combing all the feautures
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
#vectorizing the data
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
feature_vectors_fit = pd.DataFrame(feature_vectors.toarray(), columns=vectorizer.get_feature_names_out())
print(feature_vectors_fit)
```

Cosine Similarity

```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
#input from user
movie_name = input('Enter the movie name: ')
```

```python
#coverting titles to a list
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding closest match
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
#closest match
closest_match = find_close_match[0]
print(closest_match)
```

```python
#finding index of the movie
index_of_the_movie = movies_data[movies_data.title == closest_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#find the similarity score
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
#sorting mivies acc to similarity scores
sorted_similarity = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similarity)
```

```python
#finding the name of the similar movies
i = 1
for movie in sorted_similarity:
  title_of_movie = movies_data[movies_data.index == movie[0]]['title'].values[0]
  if (i<=10):
    print(i,'.',title_of_movie)
    i+=1
```

MOVIE RECOMENDATION SYSTEM

```python
movie_name = input('Enter the movie name: ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

closest_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == closest_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similarity = sorted(similarity_score, key = lambda x:x[1], reverse = True)

i = 1
for movie in sorted_similarity:
  title_of_movie = movies_data[movies_data.index == movie[0]]['title'].values[0]
  if (i<=10):
    print(i,'.',title_of_movie)
    i+=1
```
